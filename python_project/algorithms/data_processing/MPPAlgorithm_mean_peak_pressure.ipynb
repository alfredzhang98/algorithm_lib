```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 定义Mean Peak算法函数
def mean_peak_algorithm(data, threshold_factor=1.5, window_size=20):    
    peaks = []
    window = []
    mean = 0
    std = 0

    for i in range(len(data)):
        window.append(data[i])
        if len(window) >= window_size:
            mean = np.mean(window)
            std = np.std(window)
            threshold = mean + threshold_factor * std
            if data[i] > threshold:
                peaks.append(i)
            window.pop(0)
            
    return peaks
```

```python
data = np.random.randn(100)
peaks = mean_peak_algorithm(data)
print(peaks)
```

```python
fig, ax = plt.subplots()
ax.plot(data, label='Data')
ax.plot(peaks, data[peaks], 'ro', label='Peaks')
ax.legend()
plt.show()
```
